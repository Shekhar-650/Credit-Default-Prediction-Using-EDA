## TASK 4: DATA CLEANING

```python
# Checking numeric columns for data cleaning
df_app.describe()
```

Observation:

The columns such as DAYS_BIRTH, DAYS_EMPLOYED, DAYS_REGISTRATION, DAYS_ID_PUBLISH, DAYS_LAST_PHONE_CHANGE are having negative values so we need to convert it in to positive since days cannot count in negative.

DAYS_BIRTH, DAYS_EMPLOYED column can be converted to Age and YEARS EMPLOYED column in years.

values of Columns such as  AMT_INCOME_TOTAL, AMT_CREDIT, AMT_GOODS_PRICE are very high, thus we can create categorical columns by binning the values in groups.


#### Dealing with Days_columns which contains negative value and convert into positive

```python
# store all days columns which contains negative value in new variable negative_days_value_col
negative_days_value_col =["DAYS_BIRTH", "DAYS_EMPLOYED", "DAYS_REGISTRATION", "DAYS_ID_PUBLISH",'DAYS_LAST_PHONE_CHANGE']
negative_days_value_col
```

```python
# Convert ll days columns which contains negative values to positive using absolute function
df_app[negative_days_value_col]= abs(df_app[negative_days_value_col])
```

```python
# Check the details of numeric columns after converting to positive
df_app[negative_days_value_col].describe()
```

#### Dealing with DAYS_BIRTH, DAYS_EMPLOYEDs: Converting to years for better understanding 

```python
# Dividing the age column with 365 to convert days to years and binning and creating a categorical column for Agegroup
df_app.AGE = df_app["DAYS_BIRTH"]/365
bins = [0,20,25,30,35,40,45,50,55,60,100]
slots = ["0-20","20-25","25-30","30-35","35-40","40-45","45-50","50-55","55-60","60 Above"]
df_app['AGE_GROUP'] = pd.cut(df_app.AGE, bins=bins, labels=slots)
df_app.AGE_GROUP.value_counts()
```

```python
# Dividing the EMPLOYEMENT_YEARS column with 365 to convert days to years and binning and creating a categorical column for EMPLOYEMENT_YEARS
df_app.YEARS_EMPLOYED = df_app.DAYS_EMPLOYED/365
bins = [0,5,10,15,20,25,30,10000]
slots = ["0-5","5-10","10-15","15-20","20-25","25-30","30 Above"]
df_app['EMPLOYEMENT_YEARS'] = pd.cut(df_app.YEARS_EMPLOYED, bins=bins, labels=slots)
df_app.EMPLOYEMENT_YEARS.value_counts()
```

```python
df_app.YEARS_EMPLOYED.isnull().sum()
```
